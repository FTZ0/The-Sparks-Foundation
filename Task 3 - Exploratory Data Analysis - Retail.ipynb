# Exploratory Data Analysis on the given dataset ‘SampleSuperstore’.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings(action='ignore')
from pylab import rcParams
```

```python
super_store=pd.read_csv("SampleSuperstore.csv")
super_store.head()
```

# Heatmap
- Heatmap provides the correlation between every feature of the dataset.Correlation ranges from -1 to +1.

```python
sns.heatmap(super_store.isnull(), cbar=True, yticklabels=False, cmap='Reds')
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(super_store.corr(),cmap='Reds',annot=True)
```

# Boxplot 
- Boxplot is a pictorial representation of distribution. We can easily figure out outliers by using boxplots.


```python
#Boxplot of Discount
p= list(super_store.Discount) 
plt.boxplot(p) 
plt.show() 
```

```python
#Boxplot of profit
p= list(super_store.Profit) 
plt.boxplot(p) 
plt.show() 
```

```python
#Boxplot of sales
p= list(super_store.Sales) 
plt.boxplot(p) 
plt.show() 
```

# Pairplot
- Pairplot allows us to see both distribution of single variables and relationships between two variables.

```python
sns.pairplot(super_store);
```

# Correlation analysis 
-using pandas scatter_matrix()

```python
from pandas.plotting import scatter_matrix
attributes=["Profit",  "Sales"]
scatter_matrix(super_store[attributes], 
               figsize=(12,8))
```

```python
from pandas.plotting import scatter_matrix
attributes=["Quantity",  "Sales"]
scatter_matrix(super_store[attributes], 
               figsize=(12,8))
```

```python
plt.rcParams['figure.figsize'] = (15, 8)
sns.distplot(super_store['Sales'], color = 'green')
plt.title('The Distribution of Sales', fontsize = 20)
plt.xlabel('Range of Sales', fontsize = 15)
plt.ylabel('No. of Sales count', fontsize = 15)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 8)
sns.distplot(super_store['Quantity'], color = 'green')
plt.title('The Distribution of Sales Quantity', fontsize = 20)
plt.xlabel('Range of  Sales Quantity', fontsize = 15)
plt.ylabel('No. of Sales Quantity count', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('The Distribution of Sales on different regions', fontsize = 20)
plt.bar('Region','Sales',data=super_store,color='b');
```

```python
plt.figure(figsize=(15,8))
plt.title('The Distribution of Discount on sales', fontsize = 20)
plt.bar('Discount','Sales',data=super_store,color='b');
```

```python
plt.figure(figsize=(18,8))
plt.title('The Distribution of Cetagory and Sub-Cetagory', fontsize = 20)
plt.bar('Sub-Category','Category',data=super_store,color='b');
```

```python
super_store.groupby('Sub-Category')['Profit','Sales'].agg(sum).plot.bar()
plt.title('Total Profit and Sales per Sub-Category');
```

```python
super_store.groupby('State')['Profit','Sales'].agg(sum).plot.bar()
plt.title('Total Profit and Sales in differnet State');
```
